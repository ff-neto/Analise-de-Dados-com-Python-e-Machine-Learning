```python
import matplotlib.pyplot as plt
```

```python
nomes = [ 'Rafael', 'Maria', 'Ana' ]
idades = [ 10, 20, 15 ]
```

```python
# gráfico de linhas
plt.plot(nomes, idades)
```

```python
# gráfico de barras
plt.figure(figsize = ( 7, 4 ))
plt.bar(nomes, idades)
plt.xlabel('nome')
plt.ylabel('idade')
```
