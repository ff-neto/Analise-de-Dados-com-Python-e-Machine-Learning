# Feature Engineeriing 

```python
import pandas as pd
```

```python
# Lendo arquivo excel
dfClientes   = pd.read_excel('caso_estudo.xlsx', sheet_name = 'clientes')
dfLojas      = pd.read_excel('caso_estudo.xlsx', sheet_name = 'lojas')
dfProdutos   = pd.read_excel('caso_estudo.xlsx', sheet_name = 'produtos')
dfVendas     = pd.read_excel('caso_estudo.xlsx', sheet_name = 'vendas')
dfPagamentos = pd.read_excel('caso_estudo.xlsx', sheet_name = 'pagamentos')
```

```python
# Corrigindo o index
dfClientes   = dfClientes.set_index('id')
dfLojas      = dfLojas.set_index('id')
dfProdutos   = dfProdutos.set_index('id')
dfVendas     = dfVendas.set_index('id')
dfPagamentos = dfPagamentos.set_index('id')
```

```python
# Join
df = dfVendas.join( dfClientes.add_prefix('clientes_'), on = 'id_cliente' )
df = df.join( dfLojas.add_prefix('loja_'), on = 'id_loja' )
df = df.join( dfProdutos.add_prefix('produto_'), on = 'id_produto' )
df = df.join(dfPagamentos.set_index('id_venda'))
```

```python
df
```

```python
# Criando uma nova coluna se houve pagamento ou não
# Falando que tadas as vendas tiveram pagamentos
df['pagamento'] = 1
df
```

```python
# Vamos altera o pagamento para 0 onde não temos registro de data de pagamentos
# df.dt_pgto.isnull() -> Está buscando todos os registros em que a data pagamento é nula.
# --
# .loc[linhas, nome_coluna] -> As linhas são aquelas que a coluna data pagamento é nula e a coluna é o pagamento, com isso está inserindo o valor 0 nelas.
df['pagamento'] = 1
df.loc[df.dt_pgto.isnull(), 'pagamento'] = 0
df
```

```python
# Criando o tempo de pagamento
# O tempo de pagamneto é a data de pagamento - a data da venda.
# dt.days -> esse comando conveste em números inteiros
df['tempo_pg'] = (df.dt_pgto - df.dt_venda).dt.days
df
```
